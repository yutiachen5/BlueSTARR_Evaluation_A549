```python
import os
import glob
import pandas as pd

directory = '/hpc/group/igvf/A549/full-set/DMSO-200/300-bases/slurms/train-unbiased-normalized/outputs/' 

rho_list = []
mse_list = []
min_val_loss_list = []
run_list = []

for filepath in glob.glob(os.path.join(directory, '*.out')):
    with open(filepath, 'r') as file:
        run = (filepath.split(directory)[-1]).split('.out')[0]
        run_list.append(int(run))
        rho = mse = min_val_loss = None
        for line in file:
            line = line.strip()
            if line.startswith('A549 rho='):
                parts = line.split()
                rho = float(parts[2])
            elif line.startswith('A549 mse= '):
                mse = float(line.split('=')[1])
            elif line.startswith('Min validation loss:'):
                min_val_loss = float(line.split(':')[1])
        
        # Only add to list if all values are found
        if rho is not None and mse is not None and min_val_loss is not None:
            rho_list.append(rho)
            mse_list.append(mse)
            min_val_loss_list.append(min_val_loss)

df = pd.DataFrame({
    'run': run_list,
    'rho': rho_list,
    'mse': mse_list,
    'min_val_loss': min_val_loss_list,
})

df = df.sort_values(by=['run'])
df = df.reset_index(drop=True)
df
```
